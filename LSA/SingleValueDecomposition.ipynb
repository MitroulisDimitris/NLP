```python
import nltk 
import numpy as np 
import matplotlib.pyplot as publish_display_data
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD
import plotly.express as px

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
stops = set(stopwords.words('english'))
stops = stops | {'introduction','edition','series','application','approach','card',
                 'access','package','plus','etext','brief','vol','fundamental',
                 'guide','essential','printed','third','second','fourth','volume'}

titles = [line.rstrip() for line in open('all_book_titles.txt')]
```

```python
def my_tokenizer(s):
    s = s.lower()
    
    tokens = nltk.tokenize.word_tokenize(s)
    
    # remove short words
    tokens = [t for t in tokens if len(t) >2]
    
    # base form
    tokens = [wordnet_lemmatizer.lemmatize(t) for t in tokens]
    
    # stopwords
    tokens = [t for t in tokens if t not in stops]
    
    # digits
    tokens = [t for t in tokens if not any(d.isdigit() for d in t)]
        
    return tokens
```

```python
# binary = true as texts are short and we mostly care if they exist in a text
vectorizer = CountVectorizer(binary= True,tokenizer= my_tokenizer)

X = vectorizer.fit_transform(titles)

index_word_map = vectorizer.get_feature_names_out()

X = X.T

svd = TruncatedSVD()
Z = svd.fit_transform(X)
```

```python
fig  = px.scatter(x=Z[:,0],y=Z[:,1],text = index_word_map,size_max = 60)
fig.update_traces(textposition='top center')    
fig.show()
```
